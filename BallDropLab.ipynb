## Number of Data Points and Conditional Statements:

We can ask python to check whether the two lists are the same length. Perhaps we are collecting data and would like to make sure we have the same number of x data points as y data points. Previously we may have done this by simply using the ```len(xdata1)``` and ```len(ydata1)``` commands and then simply used our brains to compare the two numbers. Let's do this here:

As seen above, the ```==``` operator returns ```True``` if the two values are the same. As you can check below, it will return ```False``` if the two values are different.

But we can do better...

### If-Else Statements

If-else statements are a type of conditional statement. For the current purpose, you can think of them as a machine that can receive two inputs: ```True``` or ```False```.

```
if (True):
    r=6
else:
    r=789
```
The input here is the ```True``` that I've written. We can instead change the input to ```False``` for a different result. 

The above code sets the variable $r$ to be equal to 6 if ```True``` is inputted, but instead sets the variable $r$ to be equal to 789 if ```False``` is inputted. 

To check that this is indeed what the code is doing, we can try printing $r$ afterwards to confirm the result. Let's do this here:

````python
if (False):
    r=6
else:
    r=789

print(r)
````

But we can have python perform other tasks within the If-Else statatment! For instance, we can have python print stuff for us depending on the True/False input. Example: 

````python
if (True):
    print("Looks good :-)")
else:
    print("Something is wrong here!")
````

### Putting it all together!

The ```==``` operator gives us a way to evaluate statements by outputting True/False.

If-Else statements take True or False as an input, and then have python perform a task for us dependeing on the input that we gave it. We can combine these tools to yield the following result:

````python
# determine the lengths of the data lists
xdata1length = len(xdata1)
ydata1length = len(ydata1)

# print out the lengths- visually check that they are the same
print("number of data points (x):", xdata1length)
print("number of data points (y):", ydata1length)

if (xdata1length == ydata1length):
    print("Looks good :-)")
else:
    print("Something is wrong here!!!")
````

## Scatter Plots 

It is common in most fields to analyze sets of (x,y) data points. For this type of data the typical plot type is called a scatter plot which is just what you think of when you plot individual data points.

To begin the process in python we need to create a container for the multiple plots we will be creating. One way (not the only way) to dothis is with the ```plt.subplots``` function. This creates a container (called fig1 in this case) and a first set of axes called ax1 in this case. 

[pyplot subplots documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html)

We can then layer multiple plots onto these axes (ax1) by plotting and replotting until we are ready to show the whole thing. In this cell I am only creating a single plot of the first data set.

[pyplot scatter documentation](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.scatter.html) 

````python
# create a figure with a set of axes as we did with histograms
fig1, ax1 = plt.subplots()

# scatter plot data set 1
ax1.scatter(xdata1, ydata1)

# a more explicit way to set labels
plt.xlabel('independent variable (units)', fontsize = 10)
plt.ylabel('dependent variable (units)', fontsize = 10)
plt.title('data and model', fontsize = 20)

# turn on plot grid
ax1.grid()

# Set the size of my plot for better visibility
fig1.set_size_inches(10, 9)

# uncomment this line if I want to save a png of the plot for other purposes
#fig1.savefig("myplot.png")
plt.show()
````

## Line Plots

To plot a line using data instead of a scatter plot (that is, to "connect the dots", we can use ```plt.plot``` instead of ```ax1.scatter```. For example:


````python
# create a figure with a set of axes as we did with histograms
fig1, ax1 = plt.subplots()

# line plot data set 1
plt.plot(xdata1, ydata1)

# a more explicit way to set labels
plt.xlabel('independent variable (units)', fontsize = 10)
plt.ylabel('dependent variable (units)', fontsize = 10)
plt.title('data and model', fontsize = 20)

# turn on plot grid
ax1.grid()

# Set the size of my plot for better visibility
fig1.set_size_inches(10, 9)

# uncomment this line if I want to save a png of the plot for other purposes
#fig1.savefig("myplot.png")
plt.show()
````

We could even put both the dots and the lines on the same plot!

````python
# create a figure with a set of axes as we did with histograms
fig1, ax1 = plt.subplots()

# both representations of the data
ax1.scatter(xdata1, ydata1)
plt.plot(xdata1, ydata1)

# a more explicit way to set labels
plt.xlabel('independent variable (units)', fontsize = 10)
plt.ylabel('dependent variable (units)', fontsize = 10)
plt.title('data and model', fontsize = 20)

# turn on plot grid
ax1.grid()

# Set the size of my plot for better visibility
fig1.set_size_inches(10, 9)

# uncomment this line if I want to save a png of the plot for other purposes
#fig1.savefig("myplot.png")
plt.show()
````

## Plotting Custom Functions
It starts by defining a set of x values.```numpy.linspace()``` is a tool for doing this and because we did ```import numpy as np``` it shows in the code as ```np.linspace()```

[numpy.linspace documentation](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html)

What the function does is generate a list of values that are evenly distributed between 'begin' and 'end' in ```np.linspace('begin','end',# values)```  

```x = np.linspace(0,5,11)```

As a specific example the example above creates a list of 11 values equally spaced between 0 and 10, inclusive.  We can veryify this by printing x.

````python
x = np.linspace(0,5,11)
print(x)
````

The data structures created by ```np.linspace``` are slightly different than standard python lists in one key way: With ```np.linspace```, you can easily operate on all the elements of the list created by ```np.linspace``` by simply operating on the object ```x```. Two examples:



````python
y=2*x
print(y)

z=np.cos(x)
print(z)
````

Be careful. The calculation that looks like it's relating a single x and y value is really connecting a list of x and y values.

This is different from standard Python lists, for which we had to do use the following syntax:

````python
w1 = [1,2,3]
w2 = [v**2 for v in w1]
print(w2)
````

We can then use what we've learned to plot a function of our choice. For example:

````python
# Generate x values from 0 to 2*pi with small intervals
q = np.linspace(0, 2*np.pi, 12)

# Calculate the corresponding y values for cos(x)
qy = np.cos(q)

# Plot the cosine function
plt.plot(q, qy)

# Add labels and title
plt.xlabel('q')
plt.ylabel('cos(q)')
plt.title('Plot of cos(q)')

# Show the plot
plt.show()
````

We now have the tools to plot both a scatter plot list and a function on the same plot.

````python
# create a figure with a set of axes as we did with histograms
fig1, ax1 = plt.subplots()

# Generate x values for plotting the function
xmodel = np.linspace(0, 160, 1000)

# Calculate the corresponding y values for the plotted function
ymodel = (xmodel/16) + 0.5

# plotting the data. Note that I have added some features
ax1.scatter(xdata1, ydata1, marker = 'x', color = 'black', label = "Data")
# plotting the function 
plt.plot(xmodel, ymodel, color = 'red', linestyle = ':', linewidth = 3., label = "Model function")

# a more explicit way to set labels
plt.xlabel('independent variable (units)', fontsize = 10)
plt.ylabel('dependent variable (units)', fontsize = 10)
plt.title('data and model', fontsize = 20)

# turn on plot grid
ax1.grid()

# Set the size of my plot for better visibility
fig1.set_size_inches(10, 9)

# uncomment this line if I want to save a png of the plot for other purposes
#fig1.savefig("myplot.png")

#This creates the legend
plt.legend(loc= 4, fontsize = 20)

plt.show()
````
